## Hello ```numpy```

````python
import cv2

img = cv2.imread("assets/stasja_bear.png")[:, :, ::-1]
````

....and this is a numpy array. A matrix.

````python
img
````

Again, we need to think what one wants to do with matrices. A matrix is an _instrument_ that allows for faster mathematical operations. So we probably want to perform:
* mathematical operations
* looking at the matrix (duh)
* looking at specific parts of the matrix
* modifying parts of the matrix
* removing some parts
* adding new parts
* visual transformations

````python
import matplotlib.pyplot as plt

plt.imshow(img)
````

````python
plt.imshow(img[0:450])
````

````python
plt.imshow(img[0:450, 250:500])
````

````python
plt.imshow(img[0:450, 250:500, :1])
````

````python
plt.imshow(img[:, 250:500])  # means we take the entire height of the image and only 250-500 in width
````

````python
plt.imshow(img[:, ::-1])
````
